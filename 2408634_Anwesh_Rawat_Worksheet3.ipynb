## Which passenger had the highest fare paid relative to their age?

```python
titanic['fare_per_year']=titanic['Fare']/titanic['Age']
titanic.head()
```

```python
high_fare_age=titanic[(titanic['fare_per_year']>5)]
high_fare_age.head()
```

```python
high_fare_age_srt=high_fare_age.sort_values(by='fare_per_year',ascending=False)
high_fare_age_srt.head()
```

```python
result=high_fare_age_srt[['Name','fare_per_year']]
result.head()
```

**Master. Hudson Trevor Allison** had the highest fare paid relative to his age.



## Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their class?


```python
titanic['fare_per_class']=titanic['Fare']/titanic['Pclass']
titanic.head()
```

```python
adult_males=titanic[(titanic['Age']>=18) & (titanic['Sex']=='male')]
adult_males.head()
```

```python
adult_males_srt=adult_males.sort_values(by='fare_per_class',ascending=False)
adult_males_srt.head()
```

```python
result=adult_males_srt[['Name','Age','fare_per_class']]
result.head()
```

**Mr.Gustave J Lesurer and Mr. Thomas Drake Martinez Cardeza** of age 35 and 36 respectively paid the highest fare relative to their class.

## 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

## What percent of the total fare revenue came from each passenger class?

```python
total_fare=titanic['Fare'].sum()
total_fare
```

```python
fclass_passengers=titanic[(titanic['Pclass']==1)]
fclass_passengers.head()
```

```python
sclass_passengers=titanic[(titanic['Pclass']==2)]
sclass_passengers.head()
```

```python
tclass_passengers=titanic[(titanic['Pclass']==3)]
tclass_passengers.head()
```

```python
fare_totals=titanic.groupby('Pclass')['Fare'].sum()
fare_totals
```

```python
print(f"First class passengers fare: {format(fare_totals[1]/total_fare*100,'.2f')}%")
print(f"Second class passengers fare: {format(fare_totals[2]/total_fare*100,'.2f')}%")
print(f"Third class passengers fare: {format(fare_totals[3]/total_fare*100,'.2f')}%")
```

## What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?


```python
titanic['age_group']=pd.cut(titanic['Age'],bins=[0,18,65,200],labels=['child','adult','senior'],right=False)#This makes range like [0,18),[18,65),[65,200) which means values from 0-17
titanic.head()
```

```python
titanic.shape
#Prints a tuple of (rows,columns)
```

```python
#By Calculating length of dataframe
total_passengers=len(titanic)
total_passengers
```

```python
total_each=titanic.groupby('age_group',observed=False)['Name'].count()
total_each
```

```python
print(f"Child passengers percentage: {format(total_each['child']/total_passengers*100,'.2f')}%")
print(f"Adult passengers percentage: {format(total_each['adult']/total_passengers*100,'.2f')}%")
print(f"Senior passengers percentage: {format(total_each['senior']/total_passengers*100,'.2f')}%")
```
