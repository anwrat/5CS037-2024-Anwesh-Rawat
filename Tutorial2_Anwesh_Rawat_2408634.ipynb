## <font color="red">Problem-1</font>

### For the following dataset construct:
* Frequency Table with Relative frequency and Cumulative frequency.
* Grouped Frequency Table using square root method to determine the class interval

### 1. The scores (out of 100) obtained by 33 students in a mathematics test are as follows: 69, 48, 84, 58, 48, 73, 83, 48, 66, 58, 84, 66, 64, 71, 64, 66, 69, 66, 83, 66, 69, 71, 81, 71, 73, 69, 66, 66, 64, 58, 64, 69, 69.

```python
scores = [69, 48, 84, 58, 48, 73, 83, 48, 66, 58, 84, 66, 64, 71, 64, 66, 69, 66, 83, 66, 69, 71, 81, 71, 73, 69, 66, 66, 64, 58, 64, 69, 69]
unique_scores = sorted(list(set(scores)))
frequency = [scores.count(score) for score in unique_scores]
relative_frequency = [f / len(scores) for f in frequency]
cumulative_frequency = np.cumsum(frequency)

freq_table = pd.DataFrame({
    'Score': unique_scores,
    'Frequency': frequency,
    'Relative Frequency': relative_frequency,
    'Cumulative Frequency': cumulative_frequency
})
print("Frequency Table:")
freq_table
```

```python
#For grouped frequency table
totalobs=len(scores)
classnumber=np.sqrt(totalobs)
max_value=max(unique_scores)
min_value=min(unique_scores)
rangee=max_value-min_value #Changing range name to avoid conflicting with range function
class_width=rangee/classnumber
class_width=int(class_width)#Making class width whole number for convenience
interval=[]
interval_start=min_value
while interval_start<max_value:
  interval.append((interval_start,interval_start+class_width))
  interval_start=interval_start+class_width

grouped_frequency=[]
for i in interval:
  count=0
  for score in scores:
    if i[0]<=score<i[1]:
      count+=1
  grouped_frequency.append(count)
grouped_freq_table = pd.DataFrame({
    'Interval': interval,
    'Frequency': grouped_frequency
})
print("\nGrouped Frequency Table:")
grouped_freq_table
```

```python
plt.hist(scores, edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Histogram for Scores')
plt.grid(True)
plt.show()
```

Slighly Left Skewed Histogram <br>
Gaps between histogram suggests mixture of distribution of marks among students

```python
plt.plot(unique_scores, frequency, marker='o')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Frequency Plot')
plt.grid(True)
plt.show()
```

```python
plt.plot(unique_scores, cumulative_frequency, marker='o')
plt.xlabel('Score')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Frequency Plot')
plt.grid(True)
plt.show()
```

## <font color="red">Problem-3</font>

```python
weight_intervals = ['0-40', '40-50', '50-60', '60-70', '70-100']
frequencies = [5, 6, 8, 4, 2]
bins = [0, 40, 50, 60, 70, 100]
class_width=[bins[i+1]-bins[i] for i in range(len(bins)-1)]
relative_freq=[i/sum(frequencies) for i in frequencies]
density_1=[relative_freq[i]/class_width[i] for i in range(len(class_width))]
plt.bar(bins[:-1], density_1, width=np.diff(bins), align='edge')
plt.xlabel('Weight Interval')
plt.ylabel('Density')
plt.title('Histogram for Weight Intervals')
plt.grid(True)
plt.show()
```

Positively Skewed Histogram, Most of the concentrations around 40-60
